```python
# Importing the needed python libraries

import pandas as pd
import os
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Concatinating all the files into one file

files =  [file for file in os.listdir('C:/Users/tkjay/Desktop/salesdata')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('C:/Users/tkjay/Desktop/salesdata/'+ file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data.head()    
all_months_data.to_csv('all_data',index = False)
```

```python
all_data = all_months_data
```

```python
all_data.dropna(inplace = True)
```

```python
all_data.isnull().sum()
```

```python
#Changing the order date into date and time format.
```

```python
all_data['Data_month'] = all_data['Data_month'].dt.date
```

```python
all_data['Data_month'] = pd.to_datetime(all_data['Order Date'],errors='coerce')
```

```python
all_data.head()
```

```python
#Changing the columns to numeric data type to facilitate arithmetic operations.
```

```python
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'], errors = 'coerce')
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'],errors = 'coerce')
```

```python
all_data.head()
```

```python
all_data['Total_Price'] = all_data['Quantity Ordered']*all_data['Price Each']
```

```python
all_data.head()
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data.head()
```

```python
all_data = all_data[all_data['Month'].str[0:2]!="Or"]
```

```python
all_data.head()
```

### The best sales and the earning per month?

```python
results = all_data.groupby('Month').sum()
results.head()
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1,13)
plt.bar(months, results['Total_Price'])
plt.xticks(months)
plt.ylabel('Total_Sales in ($)')
plt.xlabel('Months')
plt.show()
```

### Which location[city] has a higher number of purchase.


```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
all_data['Purchase_city'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')
all_data.head()
```

```python
all_data.head()
```

```python
results = all_data.groupby('Purchase_city').sum()
results.head()
```

```python
cities = [city for city,df in all_data.groupby('Purchase_city')]

plt.bar(cities,results['Total_Price'])
plt.xlabel('City')
plt.ylabel('Sales in ($)')
plt.xticks(cities, rotation = 'vertical',size = 8)
plt.show()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df.head(10)
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ',' .join(x))
df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
df
```

```python
from itertools import combinations
from collections import Counter
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
